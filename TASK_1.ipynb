```
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"/content/API_SP.POP.TOTL_DS2_en_csv_v2_85.csv"
data = pd.read_csv(file_path, skiprows=4)

country_populations = data[['Country Name', '2020']].dropna()

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(country_populations['Country Name'], country_populations['2020'])
plt.xticks(rotation=110)
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population Distribution Across Countries in 2020')
plt.tight_layout()
plt.show()
```
